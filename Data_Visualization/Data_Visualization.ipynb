#### Data Visualization in Python
Iris Dataset

```python
import pandas as pd

iris = pd.read_csv("Iris.csv")

iris.head()
```

```python
iris['Species'].value_counts()
```

```python
iris.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
sns.set(style = "white", color_codes = True)
```

```python
iris.plot(kind = "scatter", x = "SepalLengthCm", y = "SepalWidthCm")
```

```python
sns.scatterplot(x = "SepalLengthCm", y = "SepalWidthCm", data = iris)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x = "SepalLengthCm", y = "SepalWidthCm", hue = "Species", data = iris)
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(10, 8))
ax = sns.boxplot(x = "Species", y = "PetalLengthCm", data = iris)
ax = sns.stripplot(x = "Species", y = "PetalLengthCm", data = iris)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.violinplot(x = "Species", y = "PetalLengthCm", data = iris)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.kdeplot(x = "PetalLengthCm", hue = "Species", data = iris)
plt.show()
```

```python
# pairplot
sns.pairplot(iris.drop("Id", axis = 1), hue = "Species")
plt.show()
```

```python
corr = iris.corr()
corr
```

```python
sns.heatmap(corr, annot = True, cmap = "Blues")
plt.show()
```
